```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib notebook
```

```python
data = np.loadtxt('test.txt',delimiter=',')
```

```python
data.shape
```

```python
mag = (data[:,::2]**2 + data[:,1::2]**2)**0.5
phase = np.arctan(data[:,1::2]/data[:,0::2])
phase2 = np.arctan2(data[:,1::2],data[:,0::2])
```

```python
plt.figure(1)
plt.clf()
plt.plot(mag[0,:],'o-')
plt.xlabel('sample')
plt.ylabel('magnitude')
```

```python
plt.figure(2)
plt.plot(phase[-1,:],'o-')
plt.plot(phase2[-1,:],'+-')
```

```python
plt.figure(3)
idx = 0
plt.plot(data[idx,::2], data[idx,1::2],'o')
plt.axis('square')
```

```python
plt.figure(4)
ax = plt.axes(projection='3d')
xline = data[idx,::2][::2]
yline = data[idx,1::2][::2]
zline = np.arange(len(xline))
# ax.plot3D(xline, yline, zline, 'gray')
ax.scatter3D(xline, yline, zline, 'b');
xline = data[idx,::2][1::2]
yline = data[idx,1::2][1::2]
zline = np.arange(len(xline))
# ax.plot3D(xline, yline, zline, 'gray')
ax.scatter3D(xline, yline, zline, 'r');
```
